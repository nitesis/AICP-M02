## 5. Loss and accuracy differences between training and testing

Why: These plots help to see how big the differences between the training and test values are. Large differences can indicate overfitting.
Plot: Line graph showing the difference between training and testing for losses and accuracies.

```python
import numpy as np

# Calculate the difference between train and test losses for each model
baseline_loss_diff = np.array(baseline_train_losses) - np.array(baseline_test_losses)
testB_loss_diff = np.array(testB_train_losses) - np.array(testB_test_losses)
testC_loss_diff = np.array(testC_train_losses) - np.array(testC_test_losses)
testD_loss_diff = np.array(testD_train_losses) - np.array(testD_test_losses)
testE_loss_diff = np.array(testE_train_losses) - np.array(testE_test_losses)
testF_loss_diff = np.array(testF_train_losses) - np.array(testF_test_losses)
testG_loss_diff = np.array(testG_train_losses) - np.array(testG_test_losses)

# Plot the differences
plt.plot(baseline_loss_diff, label="Baseline Loss Difference")
plt.plot(testB_loss_diff, label="TestB Loss Difference")
plt.plot(testC_loss_diff, label="TestC Loss Difference")
plt.plot(testD_loss_diff, label="TestD Loss Difference")
plt.plot(testE_loss_diff, label="TestE Loss Difference")
plt.plot(testF_loss_diff, label="TestF Loss Difference")
plt.plot(testG_loss_diff, label="TestG Loss Difference")

# Labels and title
plt.xlabel("Epochs")
plt.ylabel("Loss Difference (Train - Test)")
plt.legend()
plt.title("Loss Difference between Train and Test over Epochs")
plt.show()
```
